```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('UberDataset.csv')
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(1155,axis=0,inplace=True)
```

```python
df['START_DATE']=pd.to_datetime(df['START_DATE'])
```

```python
df.info()
```

```python
df.dropna(subset=['END_DATE'], inplace=True)
```

```python
df['END_DATE']=pd.to_datetime(df['END_DATE'])
```

```python
df.info()
```

```python
from datetime import datetime

df['date']=pd.DatetimeIndex(df['START_DATE']).date
```

```python
df.head()
```

```python
df['time']=pd.DatetimeIndex(df['START_DATE']).hour
```

```python
df.info()
```

```python
df.head()
```

```python
df['day-night'] = pd.cut(x=df['time'],
                              bins = [0,10,15,19,24],
                              labels = ['Morning','Afternoon','Evening','Night'])
```

```python
df.info()
```

```python
df.head()
```

```python
# creating a new label called Duration of the journey

df['starting_hour']=pd.DatetimeIndex(df['START_DATE']).hour
df['starting_min']=pd.DatetimeIndex(df['START_DATE']).minute

df['ending_hour']=pd.DatetimeIndex(df['END_DATE']).hour
df['ending_min']=pd.DatetimeIndex(df['END_DATE']).minute
```

```python
df.head()
```

```python
df['duration_hr']=df['ending_hour']-df['starting_hour']
```

```python
df['duration_min']=df['ending_min']-df['starting_min']
```

```python
df.head()
```

```python
df['Journey_duration']=df['duration_hr']*60+df['duration_min']
```

```python
df.head()
```

```python
df.drop(columns=['starting_hour','starting_min','ending_hour','ending_min'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.drop(columns=['duration_hr','duration_min'],axis=1,inplace=True)
```
